```python
import numpy as np
import LegModel
```

```python
leg = LegModel.Leg()

theta_list = []
A_l = [[], []]
B_l = [[], []]
C_l = [[], []]
D_l = [[], []]
E   = [[], []]
F_l = [[], []]
G   = [[], []]
H_l = [[], []]
U_l = [[], []]
L_l = [[], []]
```

```python
sample_rate = 10000

for theta in np.linspace(leg.theta_0, np.deg2rad(160), sample_rate):
    leg.set_tb(theta=theta, beta=0)
    theta_list.append(theta)
    
    A_l[0].append(leg.A_l[0])
    A_l[1].append(leg.A_l[1])
    B_l[0].append(leg.B_l[0])
    B_l[1].append(leg.B_l[1])
    C_l[0].append(leg.C_l[0])
    C_l[1].append(leg.C_l[1])
    D_l[0].append(leg.D_l[0])
    D_l[1].append(leg.D_l[1])
    E[0]  .append(leg.E[0])
    E[1]  .append(leg.E[1])
    F_l[0].append(leg.F_l[0])
    F_l[1].append(leg.F_l[1])
    G[0]  .append(leg.G[0])
    G[1]  .append(leg.G[1])
    H_l[0].append(leg.H_l[0])
    H_l[1].append(leg.H_l[1])
    U_l[0].append(leg.U_l[0])
    U_l[1].append(leg.U_l[1])
    L_l[0].append(leg.L_l[0])
    L_l[1].append(leg.L_l[1])
```

```python
degree = 7
print(f'A_x = {np.polyfit(theta_list, A_l[0], degree).tolist()[::-1]}')
print(f'A_y = {np.polyfit(theta_list, A_l[1], degree).tolist()[::-1]}')
print(f'B_x = {np.polyfit(theta_list, B_l[0], degree).tolist()[::-1]}')
print(f'B_y = {np.polyfit(theta_list, B_l[1], degree).tolist()[::-1]}')
print(f'C_x = {np.polyfit(theta_list, C_l[0], degree).tolist()[::-1]}')
print(f'C_y = {np.polyfit(theta_list, C_l[1], degree).tolist()[::-1]}')
print(f'D_x = {np.polyfit(theta_list, D_l[0], degree).tolist()[::-1]}')
print(f'D_y = {np.polyfit(theta_list, D_l[1], degree).tolist()[::-1]}')
print(f'E_x = {np.polyfit(theta_list, E[0]  , degree).tolist()[::-1]}')
print(f'E_y = {np.polyfit(theta_list, E[1]  , degree).tolist()[::-1]}')
print(f'F_x = {np.polyfit(theta_list, F_l[0], degree).tolist()[::-1]}')
print(f'F_y = {np.polyfit(theta_list, F_l[1], degree).tolist()[::-1]}')
print(f'G_x = {np.polyfit(theta_list, G[0]  , degree).tolist()[::-1]}')
print(f'G_y = {np.polyfit(theta_list, G[1]  , degree).tolist()[::-1]}')
print(f'H_x = {np.polyfit(theta_list, H_l[0], degree).tolist()[::-1]}')
print(f'H_y = {np.polyfit(theta_list, H_l[1], degree).tolist()[::-1]}')
print(f'U_x = {np.polyfit(theta_list, U_l[0], degree).tolist()[::-1]}')
print(f'U_y = {np.polyfit(theta_list, U_l[1], degree).tolist()[::-1]}')
print(f'L_x = {np.polyfit(theta_list, L_l[0], degree).tolist()[::-1]}')
print(f'L_y = {np.polyfit(theta_list, L_l[1], degree).tolist()[::-1]}')
```
