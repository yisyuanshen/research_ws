```python
import numpy as np
import LegModel
```

```python
leg = LegModel.Leg()

theta_list = []
A_r = [[], []]
B_r = [[], []]
C_r = [[], []]
D_r = [[], []]
E   = [[], []]
F_r = [[], []]
G   = [[], []]
H_r = [[], []]
U_r = [[], []]
L_r = [[], []]
```

```python
sample_rate = 10000  # data per deg

for rad in np.linspace(leg.theta_0, np.deg2rad(160), sample_rate):
    theta = rad
    beta = 0
    
    leg.set_tb(theta = theta, beta=beta)
    
    theta_list.append(theta)
    
    A_r[0].append(-leg.A_r[0])
    A_r[1].append(leg.A_r[1])
    B_r[0].append(-leg.B_r[0])
    B_r[1].append(leg.B_r[1])
    C_r[0].append(-leg.C_r[0])
    C_r[1].append(leg.C_r[1])
    D_r[0].append(-leg.D_r[0])
    D_r[1].append(leg.D_r[1])
    E[0].append(-leg.E[0])
    E[1].append(leg.E[1])
    F_r[0].append(-leg.F_r[0])
    F_r[1].append(leg.F_r[1])
    G[0].append(-leg.G[0])
    G[1].append(leg.G[1])
    H_r[0].append(-leg.H_r[0])
    H_r[1].append(leg.H_r[1])
    U_r[0].append(-leg.U_r[0])
    U_r[1].append(leg.U_r[1])
    L_r[0].append(-leg.L_r[0])
    L_r[1].append(leg.L_r[1])
```

```python
degree = 7
print(f'A_x = {np.polyfit(theta_list, A_r[0], degree).tolist()[::-1]}')
print(f'A_y = {np.polyfit(theta_list, A_r[1], degree).tolist()[::-1]}')
print(f'B_x = {np.polyfit(theta_list, B_r[0], degree).tolist()[::-1]}')
print(f'B_y = {np.polyfit(theta_list, B_r[1], degree).tolist()[::-1]}')
print(f'C_x = {np.polyfit(theta_list, C_r[0], degree).tolist()[::-1]}')
print(f'C_y = {np.polyfit(theta_list, C_r[1], degree).tolist()[::-1]}')
print(f'D_x = {np.polyfit(theta_list, D_r[0], degree).tolist()[::-1]}')
print(f'D_y = {np.polyfit(theta_list, D_r[1], degree).tolist()[::-1]}')
print(f'E_x = {np.polyfit(theta_list, E[0]  , degree).tolist()[::-1]}')
print(f'E_y = {np.polyfit(theta_list, E[1]  , degree).tolist()[::-1]}')
print(f'F_x = {np.polyfit(theta_list, F_r[0], degree).tolist()[::-1]}')
print(f'F_y = {np.polyfit(theta_list, F_r[1], degree).tolist()[::-1]}')
print(f'G_x = {np.polyfit(theta_list, G[0]  , degree).tolist()[::-1]}')
print(f'G_y = {np.polyfit(theta_list, G[1]  , degree).tolist()[::-1]}')
print(f'H_x = {np.polyfit(theta_list, H_r[0], degree).tolist()[::-1]}')
print(f'H_y = {np.polyfit(theta_list, H_r[1], degree).tolist()[::-1]}')
print(f'U_x = {np.polyfit(theta_list, U_r[0], degree).tolist()[::-1]}')
print(f'U_y = {np.polyfit(theta_list, U_r[1], degree).tolist()[::-1]}')
print(f'L_x = {np.polyfit(theta_list, L_r[0], degree).tolist()[::-1]}')
print(f'L_y = {np.polyfit(theta_list, L_r[1], degree).tolist()[::-1]}')
```
